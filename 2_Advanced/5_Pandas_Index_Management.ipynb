```python
# importing libraries
import pandas as pd 
from datasets import load_dataset
import matplotlib.pyplot as plt 

# loading data 
dataset = load_dataset('lukebarousse/data_jobs')
df= dataset['train'].to_pandas()

# Data Cleaning
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

# **Three Methods to cover**
    . df.reset_index()
    . df.set_index()
    . df.sort_index()

```python
df_usa = df[df['job_country'] == 'United States']
```

```python
df_usa.head(5)
```

```python
# reset_index
df_usa.reset_index(inplace=True)
df_usa.head(5)
```

```python
#set_index
#df_usa.set_index('job_index', inplace=True)

df_usa.reset_index(inplace=True)  # Moves index back to a column
df_usa.set_index('job_index', inplace=True)  # Now set it as index again
df_usa.head(5)
```

```python
print(df_usa.index.names)
```

```python
median_pivot = df_usa.pivot_table(index='job_title_short', values='salary_year_avg', aggfunc=['median', 'min', 'max'])
median_pivot
```

```python
median_pivot.sort_values(by=('median', 'salary_year_avg'), ascending=False)
```

```python
median_pivot.sort_index(inplace=True)
median_pivot
```
